```python
import json 
import pandas as pd
```

```python
#step:1 load the json category  file
files = ['IN.json', 'US.json']
cat_data=[]

for file in files:
    with open(file,'r', encoding='utf-8') as f:
        data=json.load(f)
        cat_data.append(data)
```

```python
#step:2 category mapping

cat_mapping={}
for country_data in (cat_data):
    for item in country_data['items']:
        cat_mapping[int(item['id'])]=item['snippet']['title']
```

```python
#stpe:3 load csv files and merge them
csv_files = ['in.csv', 'us.csv']
video_df_list = [pd.read_csv(file) for file in csv_files]  # load each CSV
video_df = pd.concat(video_df_list, ignore_index=True)     # merge into one
```

```python
#step:4 clean data
video_df.dropna(axis=0)
video_df.drop_duplicates()
```

```python
# Step 6: Select only required columns
video_df['category_id'] = video_df['category_id'].astype(int)
video_df['category'] = video_df['category_id'].map(cat_mapping)

columns_needed = ['publish_time', 'views', 'likes', 'comment_count', 'category_id', 'category']
video_df = video_df[columns_needed]

# Step 7: Add engagement ratio
video_df.loc[:,'engagement_ratio'] = (video_df['likes'] + video_df['comment_count']) / video_df['views']

# Step 8: Create new file with cleaned data
video_df.to_csv('clean_data.csv', index=False)
```
