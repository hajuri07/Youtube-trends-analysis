```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_df=pd.read_csv("clean_data.csv")
 
```

```python
#Top categories by Total views 

top_categories = video_df.groupby("category")["views"].sum().sort_values(ascending=False).head(10)

top_categories.plot(kind="barh", title="Top Categories by Total Views", figsize=(10,6))
plt.xlabel("Total Views")
plt.gca().invert_yaxis()

```

```python
#Average engagement ratio by category

# Group by category and calculate average engagement ratio
engagement = video_df.groupby("category")["engagement_ratio"].mean().sort_values(ascending=False)

# Plot
engagement.plot( kind="bar" ,   figsize=(10, 5),    title="Avg Engagement Ratio by Category",    color="green",    linewidth=2)
plt.ylabel("Engagement Ratio")


plt.xticks(rotation=45, ha='right') 
```
